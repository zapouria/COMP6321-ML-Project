```python
import matplotlib.pyplot as plt
import sklearn
import numpy as np   
import pandas as pd
import scipy
import sklearn.model_selection
import sklearn.neighbors
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)                          # Annoying
warnings.filterwarnings("ignore", category=sklearn.exceptions.ConvergenceWarning)  # Annoying
```

# Reading Data and Feature Extraction

```python
data= pd.read_csv('COMP6321-ML-Project-main/dataset.CSV')
y= data['Speaker'].to_numpy()
all_data = data.to_numpy()
## we will take 80/20 training testing ratio
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(all_data[:,2:], y,
                                                                            test_size=0.3, random_state=10)
#Input Normalization
scaler = sklearn.preprocessing.StandardScaler().fit(X_train)
X_train= scaler.transform(X_train)
X_test= scaler.transform(X_test)

#In this following we splitted the MFCC features and MFCC-Delta Features
MFCC_train= X_train[:,:13]
MFCC_Delta_train= X_train[:,13:]
MFCC_test= X_test[:,:13]
MFCC_Delta_test= X_test[:,13:]
```

# Classification Methods

## K nearest Neighborhood

As follows we apply the K nearest neighborhood for the MFCC data (and then for MFCC delta features), The hyperparameter in this classification are weights and k that can be changed in range of [1,40]. By considering the training and testing error, the best value for hyperparameter would be chosen.
(We prefered to use the Random hyerperarameter search).
Moreover, we use a 10 folded validation to ensure consistent distribution of classes in the training and testing phase and to also ensure that our model does not overfit for a particular split.

```python
np.random.seed(0)  
param_KNN={'n_neighbors':[np.random.randint(1, 40)], 'weights':['uniform', 'distance']} #, 
KNN= sklearn.neighbors.KNeighborsClassifier()
RandomizedSearch= sklearn.model_selection.RandomizedSearchCV(KNN,param_KNN,
                                                             random_state=0, verbose=1, n_iter=20, cv=10)
RandomizedSearch.fit(MFCC_train, y_train); 

KNN_Score= RandomizedSearch.best_score_*100
print("The best score is: %.2f" %KNN_Score)
print("The best patameters is:", RandomizedSearch.best_params_)

# Training and testing Scores
GStrain=RandomizedSearch.best_estimator_.score(MFCC_train,y_train)*100
GStest= RandomizedSearch.best_estimator_.score(MFCC_test,y_test)*100
print('Randomized search:')
print('%.1f%% train accuracy' %GStrain)
print('%.1f%% test accuracy' %GStest)
```

The results are surprisingly perfect. The Confusion matrix for this result is such as follows:

```python
fig, ax = plt.subplots(figsize=(20, 20))
sklearn.metrics.plot_confusion_matrix(RandomizedSearch.best_estimator_, MFCC_test,y_test, ax=ax)
plt.show()
```

Because the data is voice, and the features are 13 dimensions, the visulazing does not help much to understand.

## Support Vector Machine

In this Classification, the method of one vs rest is used for multiclass classification. The hyperparameter is C which is changed from log1= 0 to log100=2. The search for the best hyperparameter is randomized.

```python
np.random.seed(0)  
SVC= sklearn.svm.LinearSVC(multi_class='crammer_singer')
param_SVM= {'C':scipy.stats.reciprocal(1, 100)}
SVM_search= sklearn.model_selection.RandomizedSearchCV(SVC, param_SVM,
                                                               random_state=0, verbose=1, n_iter=20, cv=10)
SVM_search.fit(MFCC_train, y_train);

SVC_Score= SVM_search.best_score_*100
print("The best score is: %.2f" %SVC_Score)
print("The best patameters is:", SVM_search.best_params_)

# Training and testing Scores
SVCtrain=SVM_search.best_estimator_.score(MFCC_train,y_train)*100
SVCtest= SVM_search.best_estimator_.score(MFCC_test,y_test)*100
print('Randomized search:')
print('%.1f%% train accuracy' %SVCtrain)
print('%.1f%% test accuracy' %SVCtest)
```

SVM also shows a good score. Let see the Confusion Matrix:

```python
fig, ax = plt.subplots(figsize=(20, 20))
sklearn.metrics.plot_confusion_matrix(SVM_search.best_estimator_, MFCC_test,y_test, ax=ax)
plt.show()
```

Lets compare F1 score. 

```python
SVM_F1_metric= sklearn.metrics.f1_score(y_test, SVM_search.best_estimator_.predict(MFCC_test), average='macro')
KNN_F1_metric= sklearn.metrics.f1_score(y_test, RandomizedSearch.best_estimator_.predict(MFCC_test), average='macro')
print('SVM_F1_metric:', SVM_F1_metric)
print('KNN_F1_metric:', KNN_F1_metric)
```

Let see the Correctly Classified Voices:

```python
predict_SVM= SVM_search.best_estimator_.predict(MFCC_test)
Correct_SVM=0
for real, predict  in list(zip(y_test, predict_SVM)):
    if real== predict:
        Correct_SVM += 1
print('%d from %d test data classified correctly' %(Correct_SVM, len(MFCC_test))) 
```

The MFCC_Delta features are much mode than MFCC features, so this make a overfitting. As we saw in slide 12, when the dimension increases, the overfitting probablity will raise, so we need to use dimension reductions using PCA. 

## Lets see if we have this problem with SVM too:

## Dimesioanlity Reduction of MFCC Delta Features

```python
X_Delta= sklearn.decomposition.PCA(n_components= 12, random_state= 0).fit(X_train)
X_Delta_train= X_Delta.transform(X_train)
X_Delta_test= X_Delta.transform(X_test)
MFCC_Delta_train.shape
```

```python
Search= sklearn.model_selection.RandomizedSearchCV(KNN,param_KNN,
                                                             random_state=0, verbose=1, n_iter=20, cv=10)
Search.fit(X_Delta_train, y_train); 

KNN_Score= Search.best_score_*100
print("The best score is: %.2f" %KNN_Score)
print("The best patameters is:", Search.best_params_)

# Training and testing Scores
GStrain=Search.best_estimator_.score(X_Delta_train,y_train)*100
GStest= Search.best_estimator_.score(X_Delta_test,y_test)*100
print('Randomized search:')
print('%.1f%% train accuracy' %GStrain)
print('%.1f%% test accuracy' %GStest)
```

Using PCA to Visualizing our Data

```python
# The original data set is 13 dimension, for visualization, it should be reduced in two dimension
# we prefer to feed the data in a DataFrame and then use the seaborn library to show the data
feat_cols = [ str(i) for i in range(MFCC_train.shape[1]) ] 
df = pd.DataFrame(MFCC_train,columns=feat_cols)
df['y'] = y_train
df['label'] = df['y'].apply(lambda i: str(i))


twoD= sklearn.decomposition.PCA(n_components= 2).fit_transform(MFCC_train)
twoD.shape
plt.figure(figsize=(16,10))
ax= sns.scatterplot(
    x=twoD[:,0], y=twoD[:,1],
    hue="y",
    palette=sns.color_palette("hls", 19),
    data=df,
    legend="full",
    alpha=0.3
)
ax.set(xlabel="First Dimension", ylabel = "Second Dimension")
ax.set_title('Visualization of MFCC features in two dimensions')
```

So, it does make sense a little.
